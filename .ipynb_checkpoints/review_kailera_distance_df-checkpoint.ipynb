```python
import pandas as pd

# replace with your file path
df = pd.read_csv("distance_df_kai.csv")

```

```python
df_tirz_gip = df[df.dock == "tirz_GIP_7BRT"]
df_tirz_gip[df_tirz_gip.distance < 4][["atom2_residue", "atom2_residue_num"]].drop_duplicates().sort_values("atom2_residue_num").reset_index(drop=True)
```

```python
df_tirz_glp1 = df[df.dock == "tirz_GLP_6XOX"]
df_tirz_glp1[df_tirz_glp1.distance < 4][["atom2_residue", "atom2_residue_num"]].drop_duplicates().sort_values("atom2_residue_num").reset_index(drop=True)
```

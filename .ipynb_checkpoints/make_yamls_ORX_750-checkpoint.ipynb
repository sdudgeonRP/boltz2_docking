```python
import pandas as pd
import yaml
import os

# Load CSV file
input_csv = 'smiles_predictions_ORX_750.csv'  # Change this to your actual filename
df = pd.read_csv(input_csv)

# Protein sequences
sequence_5wqc = "DYKDDDDAMGTKLEDSPPCRNWSSASELNETQEPFLNPTDYDDEEFLRYLWREYLHPKEYEWVLIAGYIIVFVVALIGNVLVCVAVWKNHHMRTVTNYFIVNLSLADVLVTITCLPATLVVDITETWFFGQSLCKVIPYLQTVSVSVSVLTLSCIALDRWYAICHPLMFKSTAKRARNSIVIIWIVSCIIMIPQAIVMECSTVFPGLANKTTLFTVCDERWGGEIYPKMYHICFFLVTYMAPLCLMVLAYLQIFRKLWCRQGIDCSFWNESYLTGSRDERKKSLLSKFGMDEGVTFMFIGRFDRGQKGVDVLLKAIEILSSKKEFQEMRFIIIGKGDPELEGWARSLEEKHGNVKVITEMLSREFVRELYGSVDFVIIPSYFEPFGLVALEAMCLGAIPIASAVGGLRDIITNETGILVKAGDPGELANAILKALELSRSDLSKFRENCKKRAMSFSKQIRARRKTARMLMVVLLVFAICYLPISILNVLKRVFGMFAHTEDRETVYAWFTFSHWLVYANSAANPIIYNFLSGKFREEFKAAFSCCCLGVHHHHHHHHHH"
sequence_o43614 = "MSGTKLEDSPPCRNWSSASELNETQEPFLNPTDYDDEEFLRYLWREYLHPKEYEWVLIAGYIIVFVVALIGNVLVCVAVWKNHHMRTVTNYFIVNLSLADVLVTITCLPATLVVDITETWFFGQSLCKVIPYLQTVSVSVSVLTLSCIALDRWYAICHPLMFKSTAKRARNSIVIIWIVSCIIMIPQAIVMECSTVFPGLANKTTLFTVCDERWGGEIYPKMYHICFFLVTYMAPLCLMVLAYLQIFRKLWCRQIPGTSSVVQRKWKPLQPVSQPRGPGQPTKSRMSAVAAEIKQIRARRKTARMLMIVLLVFAICYLPISILNVLKRVFGMFAHTEDRETVYAWFTFSHWLVYANSAANPIIYNFLSGKFREEFKAAFSCCCLGVHHRQEDRLTRGRTSTESRKSLTTQISNFDNISKLSEQVVLTSISTLPAANGAGPLQNW"

# Make output directory
output_dir = "narcolepsy_orx_750"
os.makedirs(output_dir, exist_ok=True)

for _, row in df.iterrows():
    drug = row['compound_name']
    # patent_ref = row['patent_ref_number']
    smiles = row['SMILES']
    
    # base_name = f"{drug}_{patent_ref}"
    base_name = drug
    
    # 5WQC YAML
    _5wqc_yaml = {
        'version': 1,
        'sequences': [
            {'protein': {'id': 'A', 'sequence': sequence_5wqc}},
            {'ligand': {'id': 'B', 'smiles': smiles}},
        ],
        'properties': [{'affinity': {'binder': 'B'}}]
    }

    # O43614 YAML
    _o43614_yaml = {
        'version': 1,
        'sequences': [
            {'protein': {'id': 'A', 'sequence': sequence_o43614}},
            {'ligand': {'id': 'B', 'smiles': smiles}},
        ],
        'properties': [{'affinity': {'binder': 'B'}}]
    }

    # Save YAML files
    with open(os.path.join(output_dir, f"{base_name}_5WQC.yaml"), 'w') as f:
        yaml.dump(_5wqc_yaml, f, sort_keys=False)

    with open(os.path.join(output_dir, f"{base_name}_O43614.yaml"), 'w') as f:
        yaml.dump(_o43614_yaml, f, sort_keys=False)

print(f"YAML files saved to '{output_dir}'")
```

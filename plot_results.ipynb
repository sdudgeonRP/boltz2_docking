```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('result_df_il17.csv')
```

```python
# Set up the plot style
sns.set(style="whitegrid")

# Create a bar chart for each unique 'dimer'
dimer_values = df['dimer'].unique()

for dimer in dimer_values:
    # Filter the DataFrame for the current dimer value
    dimer_df = df[df['dimer'] == dimer]

    plt.figure(figsize=(20, 6))

    # Create the bar plot with 'company' as hue and 'percent_hit' as the color map
    bar_plot = sns.barplot(
        x='dock',
        y='affinity_pred_value',
        #hue='percent_hit',
        data=dimer_df,
        dodge=True,
        #palette="coolwarm",  # Choose a color palette or use 'coolwarm'
        ci=None  # No confidence interval
    )

    # Color the bars based on 'percent_hit'
    #for i, bar in enumerate(bar_plot.patches):
    #    bar.set_facecolor(plt.cm.coolwarm(dimer_df['percent_hit'].iloc[i] / 100))  # Normalize percent_hit

    # Rotate the x-axis labels for better visibility
    plt.xticks(rotation=90, ha="right")

    # Set labels and title
    plt.title(f'Affinity Prediction Values for {dimer}', fontsize=16)
    plt.xlabel('Dock Number', fontsize=12)
    plt.ylabel('Affinity Prediction Value', fontsize=12)

    # Display the legend
    plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')

    # Adjust the layout and display the plot
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import yaml
import os

# Load CSV file
input_csv = 'smiles_predictions_2.csv'  # Change this to your actual filename
df = pd.read_csv(input_csv)

# Protein sequences
sequence_aa = "MGPNSEDKNFPRTVMVNLNIHNRNTNTNPKRSSDYYDRSTSPWNLHRNEDPERYPSVIWEAKCRHLGCINADGNVDYHMNSVPIQQEILVLRREPPHSPNSFRLEKILVSVGCTCVTPIVHHVA"
sequence_af = "EFRHDSRKIPKVGHTFFQKPESCPPVPGGSMKLDIGIINENQRVSMSRNIESRSTSPWNYTVTWDPNRYPSEVVQAQCRNLGCINAQGKEDISMNSVPIQQETLVVRRKHQGCSVSFQLEKVLVTVGCTCVTPVIHHVQ"

# Make output directory
output_dir = "il17"
os.makedirs(output_dir, exist_ok=True)

for _, row in df.iterrows():
    drug = row['company']
    patent_ref = row['patent_ref_number']
    smiles = row['SMILES']
    
    base_name = f"{drug}_{patent_ref}"
    
    # AA YAML
    aa_yaml = {
        'version': 1,
        'sequences': [
            {'protein': {'id': 'A1', 'sequence': sequence_aa}},
            {'protein': {'id': 'A2', 'sequence': sequence_aa}},
            {'ligand': {'id': 'drug', 'smiles': smiles}},
        ],
        'properties': [{'affinity': {'binder': 'drug'}}]
    }

    # AF YAML
    af_yaml = {
        'version': 1,
        'sequences': [
            {'protein': {'id': 'A1', 'sequence': sequence_aa}},
            {'protein': {'id': 'F1', 'sequence': sequence_af}},
            {'ligand': {'id': 'drug', 'smiles': smiles}},
        ],
        'properties': [{'affinity': {'binder': 'drug'}}]
    }

    # FF YAML
    ff_yaml = {
        'version': 1,
        'sequences': [
            {'protein': {'id': 'F1', 'sequence': sequence_af}},
            {'protein': {'id': 'F2', 'sequence': sequence_af}},
            {'ligand': {'id': 'drug', 'smiles': smiles}},
        ],
        'properties': [{'affinity': {'binder': 'drug'}}]
    }

    # Save YAML files
    with open(os.path.join(output_dir, f"{base_name}_aa.yaml"), 'w') as f:
        yaml.dump(aa_yaml, f, sort_keys=False)

    with open(os.path.join(output_dir, f"{base_name}_af.yaml"), 'w') as f:
        yaml.dump(af_yaml, f, sort_keys=False)

    with open(os.path.join(output_dir, f"{base_name}_ff.yaml"), 'w') as f:
        yaml.dump(ff_yaml, f, sort_keys=False)

print(f"YAML files saved to '{output_dir}'")
```
